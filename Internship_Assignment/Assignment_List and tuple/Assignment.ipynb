```python
student_marks = [
    ['Ankit', [41, 34, 45, 55, 63]],
    ['Aravind', [42, 23, 34, 44, 53]],
    ['Lakshay', [32, 23, 13, 54, 67]],
    ['Gyan', [23, 82, 23, 63, 34]],
    ['Pranav', [21, 23, 25, 56, 56]]
]

print(student_marks)
```

```python
print(student_marks[0][1][1])
```

```python
for student in student_marks:
    print(student[0], student[1][1])
```

# Who scored the highest marks in the subject B?

```python
highest_marks = -1
top_student = ""

for student in student_marks:
    b_marks = student[1][1]

    if b_marks > highest_marks:
        highest_marks = b_marks
        top_student = student[0]

print("Highest marks in Subject B:", highest_marks)
print("Top student:", top_student)
```

# What is the average marks scored in the subject C?

```python
total = 0

for student in student_marks:
    total = total + student[1][2]

average = total / len(student_marks)

print("Average marks in Subject C:", average)
```

# Who scored the highest percentage of marks?

```python
highest = 0
name = ""
for i in student_marks:
    student_name = i[0]
    marks_list = i[1]
    total_marks = sum(marks_list)
    percentage = (total_marks / 500) * 100
    if percentage > highest:
        highest = percentage
        name = student_name
print("Highest Percentage =", highest)
print("Student Name =", name)
```

# If considered only top-4 subjects of a candidate, then who scored the highest percentage of marks?

```python
highest = 0
name = ""
for i in student_marks:
    student_name = i[0]
    marks_list = i[1]
    marks_list.sort()
    top_4_marks = marks_list[1:]
    total = sum(top_4_marks)
    percentage = (total / 400) * 100
    if percentage > highest:
        highest = percentage
        name = student_name
print("Highest Percentage using Top 4 Subjects =", highest)
print("Student Name =", name)
```
