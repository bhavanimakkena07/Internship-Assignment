# Problem 1: Fibonacci Series

```python
#Print first N Fibonacci numbers

n = int(input("Enter the value of N: "))

a = 0
b = 1

print("Fibonacci Series:")

for i in range(n):
    print(a, end=" ")
    c = a + b
    a = b
    b = c
```

# Problem 2: Palindrome Check

```python
# Check palindrome

value = input("Enter a string or number: ")

reverse_value = value[::-1]

if value == reverse_value:
    print("Palindrome")
else:
    print("Not a Palindrome")
```

# Problem 3: Factorial of a Number (Iteration and Recursion)

```python
# Factorial using Iteration and Recursion

def factorial_recursive(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial_recursive(n - 1)


# Input
num = int(input("Enter a number: "))

# Iteration method
fact_iter = 1
for i in range(1, num + 1):
    fact_iter = fact_iter * i

# Recursion method
fact_rec = factorial_recursive(num)

# Output
print("Factorial using Iteration =", fact_iter)
print("Factorial using Recursion =", fact_rec)
```

# Problem 4: Prime Number Check and Prime Numbers from 1 to 100

```python
# Check if a number is prime

num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print(num, "is Not a Prime Number")
            break
    else:
        print(num, "is a Prime Number")
else:
    print(num, "is Not a Prime Number")


# Print prime numbers from 1 to 100

print("\nPrime numbers between 1 and 100:")

for num in range(2, 101):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num, end=" ")
```

# Problem 5: Armstrong Number

```python
num = int(input("Enter a number: "))

temp = num
sum1 = 0

# find number of digits
digits = len(str(num))

while temp > 0:
    digit = temp % 10
    sum1 = sum1 + (digit ** digits)
    temp = temp // 10

if sum1 == num:
    print(num, "is an Armstrong Number")
else:
    print(num, "is Not an Armstrong Number")
```

# Problem 6: Reverse a Number and Find Sum of Its Digits

```python
num = int(input("Enter a number: "))

temp = num
reverse_num = 0
sum_digits = 0

while temp > 0:
    digit = temp % 10
    sum_digits = sum_digits + digit
    reverse_num = reverse_num * 10 + digit
    temp = temp // 10

print("Reversed Number =", reverse_num)
print("Sum of Digits =", sum_digits)
```

# Problem 7: Find Second Largest Number in a List

# Problem 8: Count Vowels and Consonants in a String

```python
text = input("Enter a string: ")

vowels = "aeiouAEIOU"

vowel_count = 0
consonant_count = 0

for ch in text:
    if ch.isalpha():
        if ch in vowels:
            vowel_count = vowel_count + 1
        else:
            consonant_count = consonant_count + 1

print("Vowels =", vowel_count)
print("Consonants =", consonant_count)
```

# Problem 9: Find Missing Number in a Sequence (1 to N)

```python
numbers = [1, 2, 3, 5]

n = 5

total_sum = n * (n + 1) // 2
actual_sum = sum(numbers)

missing_number = total_sum - actual_sum

print("Missing Number =", missing_number)
```

# Problem 10: Check if Two Strings are Anagrams

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")

if sorted(str1.lower()) == sorted(str2.lower()):
    print("Strings are Anagrams")
else:
    print("Strings are Not Anagrams")
```
