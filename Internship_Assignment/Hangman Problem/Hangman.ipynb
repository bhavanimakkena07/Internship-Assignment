# Hangman Problem

```python
import random
def choose_word(word_list):
    word = random.choice(word_list)
    return word.upper()
def display_board(missed_letters, correct_letters, secret_word):
    print("\nMissed letters:", missed_letters)
    print("\nCurrent guess:")
    current_word = ""
    for letter in secret_word:
        if letter in correct_letters:
            current_word = current_word + letter + " "
        else:
            current_word = current_word + "_ "
    print(current_word)
def get_guess(already_guessed):
    while True:
        guess = input("Guess a letter: ").upper()
        if len(guess) != 1:
            print("Enter only one letter")
        elif not guess.isalpha():
            print("Enter alphabet only")
        elif guess in already_guessed:
            print("Already guessed")
        else:
            return guess
def check_game_over(correct_letters, missed_letters, secret_word):
    win = True
    for letter in secret_word:
        if letter not in correct_letters:
            win = False
    if win:
        print("\nYou Won!")
        print("Word was:", secret_word)
        return True
    if len(missed_letters) >= 7:
        print("\nYou Lost!")
        print("Word was:", secret_word)
        return True
    return False
def play_again():
    return input("\nDo you want to play again? (yes or no): ").lower().startswith('y')
def play():
    word_list = ["python", "jupyter" , "programming", "function","Analytics","Explore","Generative","Infinite","Network","Foundation"]
    secret_word = choose_word(word_list)
    missed_letters = []
    correct_letters = []
    while True:
        display_board(missed_letters, correct_letters, secret_word)
        guess = get_guess(missed_letters + correct_letters)
        if guess in secret_word:
            correct_letters.append(guess)
        else:
            missed_letters.append(guess)
        game_over = check_game_over(correct_letters, missed_letters, secret_word)
        if game_over:
            if play_again():
                missed_letters = []
                correct_letters = []
                secret_word = choose_word(word_list)
            else:
                break
```

```python
play()
```
